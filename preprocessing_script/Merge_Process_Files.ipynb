```python
# Converting object data type of start_time and end_time columns into datetime data type
master_df['started_at'] = pd.to_datetime(master_df['started_at'])
master_df['ended_at'] = pd.to_datetime(master_df['ended_at'])
master_df.head()
```

```python
# Adding new column "ride_length" (ended_at - started_at) to calculate the time travelled
master_df["ride_length"] = master_df["ended_at"] - master_df["started_at"]
```

```python
# Splitting and extracting the date from "started_at" column
master_df["start_date"] = [str(i).split(" ")[0] for i in master_df["started_at"]]
```

```python
# Converting start_date into datetime data type
master_df["start_date"] = pd.to_datetime(master_df["start_date"])
```

```python
# Finding the weekday of the date
master_df["week_day"] = master_df["start_date"].apply(lambda x: x.weekday())
# Converting weekday values into weekday
week_days_dict = {0: "Sunday", 1: "Monday", 2: "Tuesday", 3: "Wednesday", 4: "Thursday", 5: "Friday", 6: "Saturday"}
master_df["week_day"] = master_df["week_day"].apply(lambda y: week_days_dict[y])
master_df["week_day"]
```

```python
# Sorting the dataframe by start date in ascending order
master_df.sort_values(by=["started_at"], inplace=True, ascending=True)
master_df
```

```python
# Dropping all rows with any NaN values
master_df = master_df.dropna()
master_df.shape
```

```python
# Calculationg the euclidean distance between start and end point
"""
master_df["x"] = (master_df["end_lat"] - master_df["start_lat"])**2
master_df["y"] = (master_df["end_lng"] - master_df["start_lng"])**2
master_df["distance"] = (master_df["x"] + master_df["y"])**(1/2)
"""
master_df[master_df["distance"] == "Tuesday"]
```

```python
# Extracting year and month from start date column
master_df["year"] = pd.DatetimeIndex(master_df["start_date"]).year
master_df['month'] = pd.DatetimeIndex(master_df["start_date"]).month
```

```python
 # Converting object datatype to timedelta
master_df["ride_len"] = pd.to_timedelta(master_df["ride_length"])
```

```python
# Finding Unique Ride Types and creating Ride Type CSV
data = list(master_df.rideable_type.unique())
ride_type = pd.DataFrame(data, columns=['Ride_Type'])
ride_type['Id'] = [x for x in range(len(data))]
ride_type
```

```python
# Converting Object Ride Type to Int
def transform_ride_type(ride_type):
  ride_type = ride_type.lower()
  if ride_type == 'docked_bike':
    return 0
  elif ride_type == 'electric_bike':
    return 1
  elif ride_type == 'classic_bike':
    return 2
  else:
    return 3
master_df['rideable_type'] = master_df['rideable_type'].apply(transform_ride_type)
master_df[master_df['rideable_type'] == 1]
```

```python
# Find Unique Stations & creating new CSV
data_start_station = list(master_df.start_station_name.unique())
data_end_station = list(master_df.end_station_name.unique())
data_stations = data_start_station + data_end_station
stations = pd.DataFrame(data_stations, columns=['station'])
# Initializing the column station_id with 0
stations['station_id'] = 0
def get_station_id(station_name):
  if station_name in master_df['start_station_name'].unique():
    matched = list(master_df[master_df['start_station_name'] == station_name]['start_station_id'])
    print(matched[0])
    return matched[0]
  elif station_name in master_df['end_station_name'].unique():
    matched = list(master_df[master_df['end_station_name'] == station_name]['end_station_id'])
    print(matched[0])
    return matched[0]
  else:
    return -1
stations['station_id'] = stations['station'].apply(get_station_id)
stations.isnull().sum()
```

```python
# Finding Unique Rider Types and creating Rider Type CSV
data = list(master_df.member_casual.unique())
rider_type = pd.DataFrame(data, columns=['Rider_Type'])
rider_type['Id'] = [x for x in range(len(data))]
rider_type
```

```python
# Converting Object Rider Type (member_casual column) to Int
"""
def transform_rider_type(rider_type):
  rider_type = rider_type.lower()
  if rider_type == 'casual':
    return 0
  elif rider_type == 'member':
    return 1
  else:
    return -1
master_df['member_casual'] = master_df['member_casual'].apply(transform_rider_type)
"""
master_df[master_df['member_casual'] == 0]
```

```python
# Creating problem during ETL Pipeline
master_df.drop(['x','y'],axis=1,inplace=True)
master_df.shape
```

```python
master_df.drop(['distance'],axis=1,inplace=True)
```

```python
# Splitting Ride Length (TimeDelta Type) to extract number of days
def calculate_ride_days(ride_length):
  ride_days = ride_length.split(' ')[0]
  return int(ride_days)
master_df['ride_days'] = master_df['ride_len'].apply(calculate_ride_days)
```

```python
# Splitting Ride Length (TimeDelta Type) to extract ride time
def calculate_ride_time(ride_length):
  ride_time = ride_length.split(' ')[2]
  hh,mm,ss= int(ride_time.split(':')[0]),int(ride_time.split(':')[1]),int(ride_time.split(':')[2])
  ride_time = dt.time(hh, mm, ss)
  return ride_time
master_df['ride_time'] = master_df['ride_len'].apply(calculate_ride_time)
master_df.head()
```
